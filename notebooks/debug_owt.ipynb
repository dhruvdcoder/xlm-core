```python
# count the number of examples in each shard 
it_ds = ds.to_iterable_dataset(256)
#it_ds = ds
per_shard_count = {}
for i in reversed(range(it_ds.num_shards)):
    shard = it_ds.shard(it_ds.num_shards, i)
    count = sum(1 for _ in shard)
    print(f"Shard {i:4d}: {count} examples")
    per_shard_count[i] = count
```
